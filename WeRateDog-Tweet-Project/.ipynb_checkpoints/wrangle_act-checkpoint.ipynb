# WeRateDogs Tweets

WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog. These ratings almost always have a denominator of 10. The numerators, though? Almost always greater than 10. 11/10, 12/10, 13/10, etc. Why? Because "they're good dogs Brent." WeRateDogs has over 9 million followers and has received international media coverage.

## Gather

```python
 #Import all the necessary library
import pandas as pd
import requests
import tweepy
import json
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
response = requests.get(url)

# Save tsv to file

with open("image_predictions.tsv", mode='wb') as file:
    file.write(response.content)
```

```python
#Import image prediction file into dataframe
img_pred = pd.read_csv('image_predictions.tsv', sep='\t')
```

```python
#import  twitter archived enhanced into dataframe
twitter_arch = pd.read_csv('twitter-archive-enhanced.csv')
```

```python
consumer_key = 'Hidden'
consumer_secret = 'Hidden'
access_token = 'Hidden'
access_secret = 'Hidden'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
tweet_errors = {}
#with tweet_json.txt open as outfile and created if not available
with open('tweet_json.txt', 'w') as outfile: 
#loop through the twitter_arch for each tweet_id
    for tweet_id in twitter_arch.tweet_id:
#Use try block to catch exception
        try:
            tweet = api.get_status(tweet_id,tweet_mode='extended')
#dump each json to the text file line by line
            json.dump(tweet._json,outfile)
# Creat a new line after each dump
            outfile.write('\n')
        except Exception as e:
            print(tweet_id,"Failed",e)
# Exception that occured are stored in tweet_errors dictionary with corresponding tweet_id
            tweet_errors[tweet_id] = e
```

```python
df_list=[]
#Loop through each line in the text file and read
for line in open('tweet_json.txt', 'r'):
#Get the data as json for easily interpretation as dictionary in python
    data = json.loads(line)
    tweet_id = data['id']
    retweet_count = data['retweet_count']
    favorite_count = data['favorite_count']
#Get the tweet_id, retweet_count and favorite_count and append it a list
    df_list.append({'tweet_id': tweet_id,
                        'retweet_count': retweet_count,
                        'likes_count': favorite_count})
# The list created is converted to a dataframe    
twitter_api = pd.DataFrame(df_list, columns = ['tweet_id', 'retweet_count', 'likes_count'])
```

`twitter_arch` columns:
- **tweet_id**: The integer representation of the unique identifier for this Tweet. tweet_id is the last part of the tweet URL after "status/" → https://twitter.com/dog_rates/status/889531135344209921.
- **in_reply_to_status_id**:  If the represented Tweet is a reply, this field will contain the integer representation of the original Tweet’s ID.
- **in_reply_to_user_id**: If the represented Tweet is a reply, this field will contain the integer representation of the original Tweet’s author ID. This will not necessarily always be the user directly mentioned in the Tweet. 
- **timestamp**: UTC time when this Tweet was created.
- **source**: Utility used to post the Tweet, as an HTML-formatted string. Tweets from the Twitter website have a source value of web
- **text**: The actual UTF-8 text of the status update.The text that comes along with the tweet.
- **retweeted_status_id**: If the represented Tweet is a retweet, this field will contain the integer representation of the original Tweet’s ID.
- **retweeted_status_user_id**: If the represented Tweet is a retweet, this field will contain the integer representation of the original Tweet’s author ID.
- **retweeted_status_timestamp**: If the represented Tweet is a retweet, this field will contain the UTC time when this tweet was retweeted
- **expanded_urls**: the urls of the tweet
- **rating_numerator**: numerator value of the Dog rating,which is almost always greater than 10
- **rating_denominator**: denominator value of the dog rating which alway 10
- **name**: name of the dog
- **doggo**,**floofer**, **pupper** and **puppo**: these are the different dog stage

```python
# Display the img_pred table
img_pred
```

`img_pred` columns:
- **tweet_id**: The integer representation of the unique identifier for each tweet. tweet_id is the last part of the tweet URL after "status/"
- **jpg_url**: The url of the image of the dog
- **img_num**: The number of image tweeted for each dog
- **p1**: p1 is the algorithm's #1 prediction for the image in the tweet
- **p1_conf**: p1_conf is how confident the algorithm is in its #1 prediction(percentage) 
- **p1_dog**: p1_dog is whether or not the #1 prediction is a breed of dog(True or False)
- **p2**: p2 is the algorithm's second most likely prediction
- **p2_conf**: p2_conf is how confident the algorithm is in its #2 prediction(percentage)
- **p2_dog**: p2_dog is whether or not the #2 prediction is a breed of dog(True or False)
- **p3**: p3 is the algorithm's third most likely prediction
- **p3_conf**: p3_conf is how confident the algorithm is in its #3 prediction(percentage)
- **p3_dog**: p3_dog is whether or not the #3 prediction is a breed of dog(True or False)

## Clean

```python
#Make copies of all the different tables
twitter_arch_clean = twitter_arch.copy()
img_pred_clean = img_pred.copy()
twitter_api_clean = twitter_api.copy()
```

### Define
Remove all the error tweets from twitter_arch_clean and merge it with the twitter_api_clean

### Code

```python
# Select the index of the tweet with error
indexes = twitter_arch_clean[twitter_arch_clean['tweet_id'].isin(list(tweet_errors.keys()))].index
# Drop the tweets with error
twitter_arch_clean=twitter_arch_clean.drop(indexes)
# Combine twitter_arch and twitter_api
twitter_arch_clean = pd.merge(twitter_arch_clean, twitter_api_clean, on='tweet_id')
```

### Test

```python
twitter_arch_clean.sample(5)
```

### Define
Select and remove all observations in the twitter_arch_clean table in which in_reply_to_status_id or retweeted_status_id is present. Drop all the unwanted columns that includes, in_reply_to_status_id, retweeted_status_id, retweeted_status_user_id, retweeted_status_timestamp and in_reply_to_user_id

### Code

```python
# Select and remove all  retweeted tweets and replied retweetd
twitter_arch_clean = twitter_arch_clean[(twitter_arch_clean['in_reply_to_status_id'].isna())  &
                                         (twitter_arch_clean['retweeted_status_id'].isna())]
#Drop all the unwanted columns
twitter_arch_clean = twitter_arch_clean.drop(['in_reply_to_status_id','in_reply_to_user_id','retweeted_status_id',
                               'retweeted_status_user_id', 'retweeted_status_timestamp'], axis = 1)
```

### Test

```python
twitter_arch_clean.head()
```

### Define
Creat new column for dog stages where doggo, floofer, pupper and puppo is equal to None and then combine all the dog stages into one column.

```python
# Function to create a new column for unnamed dog stage
def unnamed(stage):
# Check for dog stage where doggo, floofer, pupper and puppo is equal to None
    if stage.doggo=='None' and stage.floofer== 'None' and stage.pupper == 'None' and stage.puppo == 'None':
        # If yes set the value unnamed
         value = 'unnamed'
    else:
        #if no set the value None
        value ='None'
    return value
# Create the new column and store the value obtined from the function call
twitter_arch_clean['unnamed'] = twitter_arch_clean.apply(unnamed, axis=1)

# Melt the all the dog stages including the unnamed column
twitter_arch_clean = pd.melt(twitter_arch_clean, id_vars=['tweet_id','timestamp','source', 'text', 'expanded_urls', 'rating_numerator',
                                                           'rating_denominator', 'name','retweet_count', 'likes_count'],
                           var_name='stages',value_name = 'value', value_vars = ['doggo', 'floofer', 'pupper', 'puppo','unnamed'])

# Drop observation where value is equal to None
twitter_arch_clean = twitter_arch_clean[twitter_arch_clean['value']!= 'None']
 
#Drop the value columns
twitter_arch_clean = twitter_arch_clean.drop('value', axis=1)
    
```

```python
twitter_arch_clean.sample(5)
```

### Define
Extract new rating_numerator and rating_denominator to maintain the decimal values of the rating_numerator

### Code

```python
# Use regrex to extract fraction from the text and partion it into rating numerator and denominator
twitter_arch_clean['rating_numerator']=twitter_arch_clean['text'].str.extract('(\d+\.*\d*\/\d+)', expand=False).str.split('/').str[0]
twitter_arch_clean['rating_denominator']=twitter_arch_clean['text'].str.extract('(\d+\.*\d*\/\d+)', expand=False).str.split('/').str[1]
```

### Define
Remove all the html tags on the source column in twitter_arch_clean table and keep the content of the html tag

### Code

```python
# Create a new list called text
text = []
# Loop through the source column
for line in twitter_arch_clean.source:
    #find the link tag (a) and extract it content
    sor = bs(line).find('a').contents[0]
    #Append the content to the text list
    text.append(sor)
# Replace the source with the new test
twitter_arch_clean['source'] = text
```

## Analysis
In this analysis section, we will be analysing data associated this tweets,dogs and prediction using numpy, pandas matplotlib and seaborn. We will answer the following question:
> 1. Which dog stage is the most liked?
> 2. Which dog stage is the most retweeted?
> 3. Which dog stage is the most rating numerator?
> 4. Most rate score value?
> 5. What is the most popular tweet source?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
tweets_df = pd.read_csv('twitter_archive_master.csv', parse_dates=['timestamp'])
tweets_df['tweet_id'] = tweets_df['tweet_id'].astype(str)
pred_df = pd.read_csv('predictions.csv')
```

```python
tweets_df.info()
```

```python
#function for bar plot
def bar_plot(data,x,y,title,xlab,ylab):
# Use seaborn barplot
    sns.barplot(data=data,x=x,y=y,ci=None);
# Add title to the plot
    plt.title(title,fontsize = 15)
# Label the x axis of the plot
    plt.xlabel(xlab,fontsize = 14)
# Label the y axis of the plot
    plt.ylabel(ylab,fontsize = 14);
```

```python
bar_plot(tweets_df,'retweet_count','stages',"The Retweet Count vs Dog Stages",'Retweet Count','Dog stages')
```

> `Puppo` stage has the highest retweet count followed by `doggo`

```python
bar_plot(tweets_df,'stages','likes_count',"The Dog Stages vs Likes Count",'Likes Count','Dog stages')
```

> `Puppo` stage has the highest likes count followed by `doggo` which is similar to retweet count

```python
tweets_df.groupby('stages')['rating_numerator'].mean().plot(kind='line',marker='*',color='r')
plt.title("Rating Numerator vs Dog Stages",fontsize = 15)
plt.xlabel('Rating Numerator',fontsize = 14)
plt.ylabel('Dog Stages',fontsize = 14);
```

> `Puppo` stage has the highest rating numerator followed by `floofer`

```python
sns.histplot(tweets_df['rating_numerator'], kde=False, bins=10, color='teal')
# Add title to the plot
plt.title("Rating Numerator Count",fontsize = 15);
```

> Most of the dogs rating is around 11

```python
fig, (ax1, ax2) = plt.subplots(1, 2,figsize = (12,5))
fig.suptitle("How numerical properties are correlated especially likes_count and retweet_count",fontsize=20)
sns.regplot(x=tweets_df['retweet_count'], y=tweets_df['likes_count'],ax=ax1)
sns.heatmap(tweets_df.corr(), annot=True,ax=ax2);
fig.tight_layout()
```

```python
ax = sns.histplot(tweets_df['source'], kde=False, bins=10,color='brown')
locs,labels = plt.xticks()
plt.title("Tweets Source Count",fontsize = 15);
#Rotate the angle of the label
plt.setp(labels, rotation=90);
```

> Majority of the tweets came from Twitter for iPhone

```python
tweets_df.nlargest(10,'rating_numerator').groupby('stages').sum()[['rating_numerator','retweet_count','likes_count']]
```

> Top 10 rating_numerator are for doggo and pupper stages and `doggo` has most retweet and likes count between them.

```python
tweets_df.nlargest(10, 'retweet_count')[['timestamp','source','rating_numerator','retweet_count','likes_count','stages']]
```
